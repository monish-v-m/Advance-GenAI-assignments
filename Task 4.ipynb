```python
#1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [50,199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        # Check minimum amount
        if amount < 50:
            print(" Minimum recharge amount is ₹50. Try again.\n")
            continue

        # Check valid plan
        if amount in valid_plans:
            print(f" Recharge successful! Plan ₹{amount} activated.\n")
            break
        else:
            print(" Invalid plan. Valid plans are 199, 299, 399, 599. Try again.\n")
validate_recharge()
```

```python
#2: Inventory Reorder Alert System

def check_inventory(stock):
    for item, qty in stock.items():
        if qty < 15:
            print(f"{item}: Reorder Alert (Stock: {qty})")
        else:
            print(f"{item}: Stock OK (Stock: {qty})")

# example dictionary
products = {
    "Sugar": 10,
    "Rice": 25,
    "Oil": 12,
    "Salt": 40
}



check_inventory(products)
```

```python
#3: Student Result Processing System

def student_result():
    marks = []   # list to store student marks
    n = int(input("Enter number of subjects: "))

    # taking input for each subject
    for i in range(n):
        score = float(input(f"Enter marks for subject {i+1}: "))
        marks.append(score)

    # calculating total using loop
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    # deciding pass or fail
    if average >= 50:
        return f"Pass (Average: {average:.2f})"
    else:
        return f"Fail (Average: {average:.2f})"


result = student_result()
print(result)
```

```python
# 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km = 12

    normal_fare = base_fare + (distance * per_km)

    if is_peak.lower() == "yes":
        # Add 25% peak charge
        final_fare = normal_fare * 1.25
    else:
        final_fare = normal_fare

    return final_fare


# Retry loop
def fare_system():
    while True:
        km = float(input("Enter distance (km): "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(km, peak)
        print(f"Estimated Fare: ₹{fare:.2f}\n")

        again = input("Calculate again? (yes/no): ")
        if again.lower() != "yes":
            break

fare_system()
```

```python
#5: Employee Attendance Eligibility Checker

def check_attendance():
    attendance = []   # list to store P/A entries
    days = int(input("Enter total number of days: "))

    # Taking attendance input
    for i in range(days):
        status = input(f"Day {i+1} attendance (P/A): ").upper()

        # validate input
        while status not in ["P", "A"]:
            print("Invalid input! Enter P or A only.")
            status = input(f"Day {i+1} attendance (P/A): ").upper()

        attendance.append(status)

    # Count present days using loop
    present = 0
    for entry in attendance:
        if entry == "P":
            present += 1

    # Calculate percentage
    percentage = (present / days) * 100

    # Return eligibility
    if percentage >= 75:
        return f"Eligible ({percentage:.2f}%)"
    else:
        return f"Not Eligible ({percentage:.2f}%)"


# Call the function
result = check_attendance()
print(result)
```

```python
#6: Password Strength Checker

def check_password():
    password = input("Enter your password: ")

    # Rule 1: Minimum length
    if len(password) < 8:
        return " Weak: Minimum length is 8"

    has_digit = False
    has_special = False
    special_chars = "!@#$%^&*"

    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Final checks
    if has_digit and has_special:
        return " Strong Password"
    else:
        return " Weak: Must include at least 1 digit and 1 special character (@#$)"


# Call the function
result = check_password()
print(result)
```
